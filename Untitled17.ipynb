```python
from transformers import BartForConditionalGeneration, BartTokenizer
import torch
device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn").to(device)
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")

text ="The Aston Martin Vulcan is a rare, track-only supercar limited to just 24 units, designed to showcase the brand's racing pedigree. It features a lightweight carbon fiber body with aggressive aerodynamics, powered by a naturally aspirated 7.0-liter V12 engine producing around 820 horsepower. The Vulcan is capable of exceeding 200 mph, with a 0-60 mph time of under 3 seconds. It comes with a six-speed sequential gearbox and advanced racing technologies, including adjustable aerodynamics and a race-spec suspension. Exclusively designed for the track, the Vulcan offers an unparalleled driving experience for those fortunate enough to own one."
inputs = tokenizer([text], max_length=1024, return_tensors="pt", truncation=True).to(device)
summary_ids = model.generate(inputs["input_ids"], num_beams=4, max_length=150, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(summary)
```

```python
def summrization(text):
    if len(text.split())>=1024:
        a=text.split(".")
        a1=a[:len(a)//2]
        a2=a[len(a)//2:]
        sum=summrization(".".join(a1))+summrization(".".join(a2))
        return sum
    else:
        inputs = tokenizer([text], max_length=1024, return_tensors="pt", truncation=True).to(device)
        summary_ids = model.generate(inputs["input_ids"], num_beams=16, max_length=500, early_stopping=False)
        summary1 = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary1
```
